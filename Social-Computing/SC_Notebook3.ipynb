## Identifying the country of residence

For example, an email like “1315ae6229444367968a943a219f38def9a8112d@vpn-251-169.epfl.ch” will probably correspond to a user located in Switzerland, while an email of the form “59bd0a3ff43b32849b319e645d4798d8a5d1e889@philipphauer.de” will probably correspond to a user located in Germany.

```python
def classify_email(emailId):
  split = emailId.split(".")
  last_value = split[-1]
  residence = "unknown"

  if last_value == "de":
    residence = "Germany"
  elif last_value == "ch":
    residence = "Switzerland"
  elif last_value == "jp":
    residence = "Japan"
  elif last_value == "io":
    residence = "British India"
  elif last_value == "nl":
    residence = "Netherlands"
  elif last_value == "in":
    residence = "India"
  elif last_value == "edu":
    residence = "USA"
  elif last_value == "ru":
    residence = "Russia"
  elif last_value == "cn":
    residence = "China"
  elif last_value == "fr":
    residence = "France"
  elif last_value == "be":
    residence = "Belgium"
  elif last_value == "tr":
    residence = "Turkey"
  elif last_value == "uk":
    residence = "UK"
  elif last_value == "tv":
    residence = "Tuvalu"
  elif last_value == "at":
    residence = "Austria"
  elif last_value == "mx":
    residence = "Mexico"
  elif last_value == "au":
    residence = "Australia"
  elif last_value == "pt":
    residence = "Portugal"
  elif last_value == "cz":
    residence = "Czech Republic"
  elif last_value == "It" or last_value == "it":
    residence = "Italy"
  elif last_value == "dk":
    residence = "Denmark"
  elif last_value == "cl":
    residence = "Chile"
  elif last_value == "ca":
    residence = "Canada"
  elif last_value == "fi":
    residence = "Finland"
  elif last_value == "br":
    residence = "Brazil"
  elif last_value == "co":
    residence = "Colombia"
  elif last_value == "lan":
    residence = "People's Republic of China"
  elif last_value == "tu":
    residence = "Tunisia"
  elif last_value == "zw":
    residence = "Zimbabwe"
  elif last_value == "hu":
    residence = "Hungary"
  elif last_value == "eu":
    residence = "Europe"
  elif last_value == "sc":
    residence = "Seychelles"
  elif last_value == "id":
    residence = "Indonesia"
  elif last_value == "kr":
    residence = "Korea"
  elif last_value == "lt":
    residence = "Lithuania"
  elif last_value == "us":
    residence = "USA"
  elif last_value == "pl":
    residence = "Poland"
  elif last_value == "bzh":
    residence = "Brittany"
  elif last_value == "dk":
    residence = "Denmark"
  
  return residence
```

```python
for i in range(len(df)):
  emailId = df[0][i]
  residence = classify_email(emailId)
  if residence != "unknown":
    print(residence, ":", emailId)
```
